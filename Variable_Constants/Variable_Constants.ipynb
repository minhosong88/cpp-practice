### Variables
- a memory location is associated with a variable name
- a variable is an abstraction for a memory location
   - Type  (integer, real number, string, etc.)
   - Value: content (10, 3.14, name, etc. )

- syntax of variable declaration: type name;
```cpp
int age;
double rate;
string name;
```
- **Rules**
   - must begin with a letter or underscore
   - case sensitive
   - no special character
   - reserved keywords cannot be used
- Convention
   - camel case or snake case
   - use meaningful names
   - Never use variables before initializing them
   - declare variables close to when you need them in your code

- Initialization
```cpp
int age = 21; // assignment
int age (21); // constructor initialization
int age {21}; // C++ 11 list initialization syntax ( encouraged to use )
```


### Primitive Types
- Fundamental types: character, integer (signed, unsigned), floating-point, boolean
- size and precision is often complier-dependent
#### Character
- char, char16_t(at least 16 bits)
#### integer
- short int (at least 16 bits)
- int (at least 16 bits)
- long int (at least 32 bits)
- long long int (at least 64 bits)
#### float
- float (7 decimal digits)
- double (15 decimal digits/ no less than float)
- long double (19 decimal digits / no less than double)
#### boolean
- bool (8 bits, true/false)

### Sizeof Operator

- '<'climits'>' '<'cfloat'>'

### Constants
- have names, occupy storage, usually typed
- cannot change once declared.
- Types
  - literal
  - declared
  - constant expressions
  - enumerated constants
  - defined constant

#### Literals
- x = 12; 
- \n : new line
- \r : return

#### declared constants
```cpp
const double pi {3.1415926};
```

#### Defined constants
```cpp
#define pi 3.1415926
```
- Don't use defined constants in Modern C++
